# Baseline Model
### Linear regression

```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
# Load training data for traning and validation sets
training_data = pd.read_csv('Kaggle Data/train.csv')
training_data
```

```python
# Select features and target variable
X = training_data[["GrLivArea", "OverallQual", "GarageCars"]]
y = training_data["SalePrice"]

y_log = np.log(y)
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X, y_log, test_size=0.2, random_state=42
)


model = LinearRegression()
model.fit(X_train, y_train)

y_val_pred_log = model.predict(X_val)
```

```python
rmse = np.sqrt(mean_squared_error(y_val, y_val_pred_log))
print("RMSE (validation set):", rmse)
```

```python
model.fit(X, y_log)
```

```python
# Load training data for traning and validation sets
test_data = pd.read_csv('Kaggle Data/train.csv')
test_data
```

```python
X_test = test_data[["GrLivArea", "OverallQual", "GarageCars"]]
test_pred_log = model.predict(X_test)

test_pred_prices = np.exp(test_pred_log)
test_pred_prices
```

```python
rmse = np.sqrt(mean_squared_error(test_data["SalePrice"], test_pred_prices))
print("RMSE (test set):", rmse)
```
